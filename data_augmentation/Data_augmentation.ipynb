### üìö –Ü–º–ø–æ—Ä—Ç –±—ñ–±–ª—ñ–æ—Ç–µ–∫

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

### üñºÔ∏è –ó—á–∏—Ç—É–≤–∞–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è

```python
image_path = 'cat.jpg'
img = tf.keras.utils.load_img(image_path)
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img)
# plt.axis('off')
plt.show()
```

### üê¨ –í–∏–±—ñ—Ä –∞—É–≥–º–µ–Ω—Ç–∞—Ü—ñ—ó –¥–∞–Ω–∏—Ö

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomFlip("vertical"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.9),
    tf.keras.layers.RandomTranslation(0.01, 0.05),
    tf.keras.layers.RandomContrast(0.3),
])
```

### üìä –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤

```python
plt.figure(figsize=(20, 20))
plt.subplots_adjust(wspace=0, hspace=0)

for i in range(9):
  augmented_image = data_augmentation(img_array)
  ax = plt.subplot(3, 3, i + 1)
  plt.imshow(augmented_image[0].numpy().astype('uint8'))
  # plt.axis("off")
```
